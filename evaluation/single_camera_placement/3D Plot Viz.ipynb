```python
scores = np.load("straight_road_300x300_6m_pitch_yaw_scores_0-4_two_stddev.npy")
```

```python
camera_pitches = np.arange(31, 92, 3.0)
camera_yaws = np.arange(0.0, 180, 5.0)
```

```python
fig = plt.figure()
ax = fig.add_subplot('111', projection='3d')
scores_flat = scores.ravel()
ps, ys = np.meshgrid(camera_pitches, camera_yaws)
ps_, ys_ = ps.ravel(), ys.ravel()
bottom = np.zeros_like(ps_)
width = ps[0, 1] - ps[0,0]
depth = ys[0,0] - ys[1,0]
print(ps_.shape, ys_.shape, scores.ravel().shape)
print(width, depth)
ax.bar3d(ps_, ys_, bottom, width, depth, scores.ravel(), shade=True)
ax.set_xlabel("Pitch Angle")
ax.set_ylabel("Yaw Angle")
plt.show()
```
